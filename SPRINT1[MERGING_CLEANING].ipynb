#  THE GAMES_FILE ,GAME_EVENT_FILE,GAME_LINEUP_FILE,PLAYERS_FILE,APPERANCE_FILE ARE MERGING INTO SINGLE DATA FILE

```python
df=app_file.merge(game_event_file,on=['game_id','player_id'],how='outer')
df1=df.merge(game_lineup_file,on=['game_id','player_id'],how='outer')
df2=df1.merge(games_file,on='game_id',how='outer')
df3=pd.merge(df2,players_file,on='player_id',how='outer')
df3
```

```python
# CHECKING NULL VALUES 
df3.isnull().sum()
```

```python
df3.dtypes
```

```python
# THE home_club_formation,away_club_formation HAS NO VALUES SO DROPPING THESE COLUMNS FROM DATA SET
df3.drop(columns=['home_club_formation','away_club_formation'],inplace=True)
```

```python
#SEGAREGATE THE OLUMNS INTO THE NUMERICAL  AND CATEGORICAL COLUMNS
import pandas as pd
import numpy as np

# segaregating the Numerical Columns from original data set
numerical_cols = df3.select_dtypes(include=['float64', 'int64']).columns
for col in numerical_cols:
    if df3[col].isnull().sum() > 0:
        df3[col].fillna(df3[col].median(), inplace=True)

# segaregating the Categorical Columns from original data set
categorical_cols = df3.select_dtypes(include=['object', 'category']).columns
for col in categorical_cols:
    if df3[col].isnull().sum() > 0:
        df3[col].fillna(df3[col].mode()[0], inplace=True)

# segaregating the Datetime Columns from original data set
datetime_cols = df3.select_dtypes(include=['datetime64[ns]']).columns
for col in datetime_cols:
    if df3[col].isnull().sum() > 0:
        # Replace with the median date
        df3[col].fillna(df3[col].median(), inplace=True)
```

```python
#CHECKING IF ANY DUPLICATES ARE THERE
df4=df3.duplicated()
df4
```

```python
#DROPPING THE DUPLIACTES FROM THE DATA
df5=df3.drop_duplicates()
df5
```

```python
#THE EXCEL FILE IS CONVERTED INTO CSV FILE AND NAMENG  AS FOOTBALL_DATA_MERGED_CLEANED.
df5.to_csv('Football_Data_Merged_Cleaned.csv', index=False)

print("File saved as 'Football_Data_Merged_Cleaned.csv'")
```
