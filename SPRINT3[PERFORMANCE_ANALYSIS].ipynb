# BUSINESSS OBJECTIVE QUESTIONS

#          Evaluate and analyze which player's performance has the highest impact on match outcomes based on goals and assists? 

```python
#Calculate player performance based on goals and assist
player_performance = data.groupby(['player_name_x','season'])[['goals', 'assists']].sum()
# Calculate impact score of player performance
player_performance['impact_score'] = player_performance['goals'] + player_performance['assists']

# Sort by impact score
top_players = player_performance.sort_values('impact_score', ascending=False).head(10)

# Reset index for easier manipulation
top_players = top_players.reset_index()

# Concatenate player names for visualization
top_players['player_name'] = top_players['player_name_x']
top_players['seasons']=top_players['season']

# Print top players
print(top_players[['player_name', 'impact_score','seasons']])

# Visualize the top players
plt.figure(figsize=(10, 6))
plt.bar(top_players['player_name'], top_players['impact_score'], color='pink')
plt.title('Top 10 Players by Impact Score')
plt.xlabel('Player Name(s)')
plt.ylabel('Impact Score')
plt.xticks(rotation=45)  # Rotate x-axis labels
plt.show()

```

# INTERPRETATION : So from that identified the top players performance based on the goals and asists.The aron johannnsson is the top player based on the goals and assists 

#   BUSINESS OBJECTIVE QUESTION :Which players contributed the most goals and assists  in the year of 2016?

```python
last_season = 2016
# Filter last season's data
df = data[data['season'] == last_season]
# Group by player and aggregate goals and assists
df1 = (df.groupby('player_name_x')[['goals', 'assists']].sum().sort_values(by=['goals', 'assists'], ascending=False))[:5]

# Display the result
print(df1)
df1.plot(kind='bar', stacked=False, figsize=(10, 6))
# Set plot title and labels
plt.title(f'Top 5 Players by Goals and Assists ({last_season})', fontsize=14)
plt.xlabel('Player', fontsize=12)
plt.ylabel('Count', fontsize=12)
# Show the plot
plt.xticks(rotation=45)
plt.show()
```

# INTERPRETATION: From that determined the Bobby Wood contributed the most goals and assists in 2016. 

# BUSINESS OBJECTIVE QUESTINS :  Which teams have the highest number of yellow and red cards in home club?

```python
cards = data.groupby(['home_club_name'])[['yellow_cards', 'red_cards']].sum().sort_values(by=['yellow_cards', 'red_cards'], ascending=False)[:10]
print(cards)


cards.plot(kind='bar', stacked=True, figsize=(10, 6))

# Set plot title and labels
plt.title('Top 10 Teams by Yellow and Red Cards', fontsize=14)
plt.xlabel('Team', fontsize=18)
plt.ylabel('Card Count', fontsize=16)

# Show the plot
plt.xticks(rotation=90)

plt.show()
```

 # PROBABILITY 

##   What is the probability that a player scored a goal and received a yellow card in the same match?

```python
# Filter rows where the player scored a goal and received a yellow card
scored_and_yellow = data[(data['goals'] >= 1) & (data['yellow_cards'] >=1)]

# Calculate the joint probability
joint_probability = len(scored_and_yellow) / len(data)

print(f"Joint Probability (Scored and Yellow Card): {joint_probability:.2f}")
```

# INTERPRETATION:So identified the probability  of yelllow card and goal is 0.01

# What is the probability that a player scoring a goal within atleast 30 minutes

```python
# Filter matches where home team scored more than 4 goals and players played >= 30 minutes
score = data[(data['home_club_goals'] > 1) & (data['minutes_played'] > 45)]

# Calculate the probability of matches meeting the condition
prob_team_score = len(score) / len(data)

print(f"Probability (Home Team Scored > 1 Goals and Played > 30 Minutes): {prob_team_score:.2f}")
```

# INTERPRETATION:So identified the probability of team scored more than one goals is 0.01

# Sampling, Central Limit Theorem, and Normal Distribution QuestionsÂ¶

```python
import numpy as np

# Sampling from goals scored
sample_means = []
for _ in range(1000):
    sample = np.random.choice(data['home_club_goals'], size=50, replace=True)
    sample_means.append(sample.mean())

# Plotting the sampling distribution of the mean
plt.figure(figsize=(10, 6))
sb.histplot(sample_means, kde=True)
plt.title('Sampling Distribution of Average Goals Scored')
plt.xlabel('Mean Goals Scored')
plt.ylabel('Frequency')
plt.show()
```

### Interpretation: The plot of sample means will approximate a normal distribution and also supporting the central limit theorem.
