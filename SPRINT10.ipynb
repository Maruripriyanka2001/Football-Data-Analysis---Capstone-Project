# SOLVE BUSINESS QUESTION RELATED TO THE K-MEANS CLUSTERING

#                 How can we segment players based on performance?

```python
# Step 1: Import Libraries
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import seaborn as sns



# Step 3: Data Preparation
# Select relevant columns for clustering (adjust column names as per your dataset)
columns_for_clustering = ['goals', 'assists', 'minutes_played']  # Example columns
data_for_clustering = data[columns_for_clustering]



# Scale the data
scaler = StandardScaler()
scaled_data = scaler.fit_transform(data_for_clustering)

# Step 4: Determine Optimal Number of Clusters (Elbow Method)
inertia = []
k_range = range(1, 11)

for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(scaled_data)
    inertia.append(kmeans.inertia_)

# Plot the Elbow Curve
plt.figure(figsize=(8, 5))
plt.plot(k_range, inertia, marker='o')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Inertia')
plt.title('Elbow Method for Optimal k')
plt.show()

# Step 5: Fit K-Means Model and Analyze Clusters
# Based on the elbow curve, choose the optimal number of clusters (e.g., k=3)
optimal_k = 3
kmeans = KMeans(n_clusters=optimal_k, random_state=42)
data['Cluster_Label'] = kmeans.fit_predict(scaled_data)

# Display cluster centers
print("Cluster Centers (scaled values):")
print(kmeans.cluster_centers_)

# Add cluster labels to the original data
data['Cluster'] = data['Cluster_Label']

# Step 6: Visualize Clusters
# Example: Scatter plot for goals vs assists colored by cluster
plt.figure(figsize=(8, 6))
sns.scatterplot(x='goals', y='assists', hue='Cluster', data=data, palette='viridis')
plt.title('Player Clustering: Goals vs Assists')
plt.xlabel('Goals')
plt.ylabel('Assists')
plt.legend(title='Cluster')
plt.show()

```

```python
INTERPRETTAION:The optimal number of clusters is around 3, as the curve starts to flatten after this point, indicating diminishing returns in reducing inertia with additional clusters.
```

```python
Player Clustering (Goals vs. Assists):

The scatter plot shows players clustered into three distinct groups (Cluster 0, Cluster 1, Cluster 2) based on goals and assists.
Cluster 0 (purple): Players with lower goals and assists.
Cluster 1 (green): Players with moderate assists but low goals.
Cluster 2 (yellow): Players with higher goals and assists.
This indicates that players can be segmented into performance tiers based on these metrics, which might help in tailoring strategies or identifying standout performers.
```
